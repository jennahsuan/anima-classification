```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
from fake_useragent import UserAgent
```

```python
animes_list = []

with open ("/Users/zhuangxiangan/Desktop/TM_crawling.csv", "r") as f:
    table = f.readlines()
    for row in table:
        animes_list.append(row.split(",")[1])

animes_list.pop(0)
print(animes_list)
```

```python
char_key_list = ["角色","主角""人物","登場","登场","配角","用語","用语","出場","出场"]

bg_key_list = ["背景","簡介","用語","劇情","介紹","細節","故事","大綱","世界","世界觀","简介","剧情","介绍","细节","大纲","用语","世界观","概要","內容","故事線","故事线"]

bg_content_list = []
char_content_list = []


counter = 0
for anime in animes_list:
    # counter += 1
    # print(counter,end="\r")
    # if counter == 20:
    #   break
    url = "https://zh.wikipedia.org/wiki/"
    url += anime

    # 設定header
    ua = UserAgent()
    user_agent = ua.random
    header = {'user-agent': user_agent}


    ## mw-headline
    #r = requests.get(url, headers = header)
    r = requests.get(url)
    html = r.content
    soup = BeautifulSoup(html,"html.parser")

    # print(soup.prettify())

    headline = soup.find_all(class_="mw-headline")



    # for hl in headline:
    #     parents = hl.find_parents("h2")
    #     for parent in parents:
    #         print(parent.text)

    bg_content = ""
    char_content = ""
    for hl in headline:
        parents = hl.find_parents("h2")
        for parent in parents:
            for keyword in bg_key_list:
                if keyword in parent.text:
                    # print(parent.text,":")
                    for sibling in parent.find_next_siblings():
                        if sibling.name != parent.name and "配音員" not in sibling.text:  #.name = h2, p, span
                            #print(sibling.name, ":")
                            #print("----------start----------")
                            # print(sibling.text.strip())
                            bg_content += sibling.text.strip()
                            #print("----------end------------")
                        else:
                            break
                    break
            for keyword in char_key_list:
                if keyword in parent.text:
                    # print(parent.text,":")
                    for sibling in parent.find_next_siblings():
                        if sibling.name != parent.name and "配音員" not in sibling.text:  #.name = h2, p, span
                            #print(sibling.name, ":")
                            #print("----------start----------")
                            #print(sibling.text.strip())
                            char_content += sibling.text.strip()
                            #print("----------end------------")
                        else:
                            # print('----------------------')
                            break
                    break


    bg_content_list.append(bg_content)
    char_content_list.append(char_content)



# for i in range(len(bg_content_list)):
#     print(i)
#     print(bg_content_list[i])
#     print("-------")
#     print(char_content_list[i])
#     print("=====================")

bg_dic = {"背景描述": bg_content_list, "角色描述": char_content_list}
df = pd.DataFrame(data=bg_dic)


# print(df)

df.to_excel("tm_ouput.xlsx")
```
